```python
import numpy as np
from scipy.special import gamma
from matplotlib import pyplot as plt

import torch
```

Experiment with the mean norm of a vector in $\mathbb R^n$ sampled from a standard normal distribution, under various $p$-norms and for various $n$.

```python
embedding_dim = 100
size = 10000
p = 2

sample = torch.randn((size, embedding_dim))
mean_norm = sample.norm(dim=1, p=p).mean()
mean_norm
```

```python
embedding_dim = 100
size = 1000

pls = np.linspace(1, 8, 100)
sample = np.random.default_rng().standard_normal((size, embedding_dim))
mean_norms = [np.linalg.norm(sample, axis=1, ord=p).mean() for p in pls]
plt.plot(pls, mean_norms)
plt.plot((pls[0], pls[-1]), (np.e, np.e))
plt.show()
```
